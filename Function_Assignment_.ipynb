Ques 2.	Explain the concept of function arguments and parameters in Python.

Answer:


Parameters are the variables listed inside the parentheses in the function definition. They act as placeholders for the values that the function will use when it is called.


Arguments are the actual values or data you pass to a function when you call it. These values are assigned to the corresponding parameters in the function definition.

```python
#example of Parameters and Arguments:
def greeting(name):             #here name is the parameter
  print ("Welcome to the office", name)
```

```python
greeting("Mohan")                 #here Mohan is argument
```

Types of Arguments and Parameters-

1.	Positional Arguments: Arguments are assigned to parameters based on their position in the function call.




```python
#example of Positional Arguments:
def addittion(a,b,c):           #here one extra argument is given for space in between.
  return a+b+c
```

```python
#example of Variable length Arguments:
def add_no(*args):
  a = 0
  for i in args:
      a = a+i
  return a
```

Ques 6. Explain the concept of generators in Python and how they are defined.

Answer:
a.	Generators are a type of iterable that generate values on-the-fly rather than storing them in memory. They produce items one at a time and only when requested, which can lead to significant memory savings.

b.	Generators use functions to yield values instead of returning them all at once. The yield keyword is used to produce a value and pause the function’s execution.

c.	Since generators produce values one at a time and do not store the entire sequence in memory, they are often more efficient than using lists or other data structures, especially for large or infinite sequences.


Defining Generators:

a.	Generator Function: A generator function is defined using the def keyword, but instead of using return to return a result, it uses yield to produce a value.



```python
#example of Generator Function
def countdown(n):
  for i in range(n,0,-1):
    yield n
    n -=1
for num in countdown(5):
    print (num)
```

b.	Generator Expression: A generator expression provides a concise way to create a generator without defining a separate function.

```python
squares = (x * x for x in range(10))
for square in squares:
    print(square)
```

Ques 7.	What are the advantages of using generators over regular functions?

Answer:

Some of the advantages of using generators are:

a.	Generators do not store the entire sequence in memory; they generate values on the fly, which makes them memory-efficient.

b.	Generators execute the values only when requested, which leads to performance improvements since they don’t compute all values at once.

c.	Generators provide a simpler way to write iterators. Using the yield keyword allows for the creation of an iterable object.

d.	Generators automatically retain their state between successive calls.

e.	Generators can lead to cleaner, more readable code for tasks that involve iteration or streaming data, as they encapsulate the logic of iteration in a more compact form.


```python
#example of generators

def divide(d):
  yield d%4 == 0
for div in divide(16):
  print(div)
```

Ques 7. Implement a generator function that reads a file line by line and yields each line as a string.

Answer:

```python
def read_file_line_by_line(file_path):
    try:
        file = open(file_path, 'r')
        with file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"Error: The file at {file_path} was not found.")
    except IOError:
        print(f"Error: An IOError occurred while trying to read the file at {file_path}.")
```

```python
file_path = 'example.txt'

for line in read_file_line_by_line(file_path):
    print(line)
```
